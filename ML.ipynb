```python
# Import important dependencies
import numpy as np
import pandas as pd
```

```python
# Importing LogisticRegression Model (Binary Output)
from sklearn.linear_model import LogisticRegression
```

```python
# Importing tratin test split function
from sklearn.model_selection import train_test_split
```

```python
# Importing accuracy function
from sklearn.metrics import accuracy_score
```

```python
# To convert All String format into numeric
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Importing Dataset
df=pd.read_csv('mail_data.csv')
df
```

```python
# Information of the dataset
df.info()
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Basic Anayalysis
# Check the ham/spam emails count
df['Category'].value_counts().reset_index()
```

```python
# Checking unique values from category column
pd.unique(df['Category'])
```

```python
# Label Encoding
# Labelling spam email as 0 and ham email as 1
df.loc[df['Category']=='ham','Category']=1
df.loc[df['Category']=='spam','Category']=0
```

```python
# Separating the mail data into Input data and Label(Output data)
x=df['Message']
y=df['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting the data into taraining and tessting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x.shape)
print(x_train.shape)
print(x_test.shape)

# Training Data - 80%
# Testing Data - 20%
```

```python
# Transforming the text into numeric machine understanding data
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
x_train_num=feature_extraction.fit_transform(x_train)
x_test_num=feature_extraction.transform(x_test)
```

```python
# Convert y_train and y_test values as integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
# Training the model
model=LogisticRegression()
```

```python
# Train the Logistic Regression model with training data
model.fit(x_train_num,y_train)
```

```python
# Evaluating the trained model
x_train_prediction=model.predict(x_train_num)
# Finding the accuracy score of training data
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy score of training data is',training_data_accuracy)
```

```python
# Evaluating the testing model
x_test_prediction=model.predict(x_test_num)
# Finding the accuracy score of testing data
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy score of testing data is',testing_data_accuracy)
```

```python
# Building the predictive system
input_data=['Congratulation u have won a lottery prixe of 2000']
input_data_num=feature_extraction.transform(input_data)
```

```python
# Making Prediction
prediction=model.predict(input_data_num)
if prediction==1:
  print("Ham Email")
else:
  print("Spam Email")
```

```python
df=pd.read_csv('car data.csv')
df
```

```python
# Basic Analysis
pd.unique(df['Fuel_Type'])
```

```python
pd.unique(df['Seller_Type'])
```

```python
# Encoding the Fuel Type column
df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}}, inplace=True)
```

```python
# Encoding the Seller Type column
df.replace({'Seller_Type':{'Dealer':0,'Individual':1}}, inplace=True)
```

```python
# Encoding the Transmission column
df.replace({'Transmission':{'Manual':0,'Automatic':1}}, inplace=True)
```

```python
# splitting the data into Input feature and label
x=df.drop(['Car_Name','Selling_Price'],axis=1)
y=df['Selling_Price']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# Prediction for training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=metrics.r2_score(x_train_prediction,y_train)
print("Accuracy Score of training Data:",training_data_accuracy)
```

```python
# Prediction of the testing Data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=metrics.r2_score(x_test_prediction,y_test)
print("Accuracy Score of testing data:",testing_data_accuracy)
```

```python
# Building Predictive System
input_data=(2014,5.59,27000,0,0,0,0)
input_data_as_numpy_array=np.asarray(input_data)
# Reshape to one data point
input_data_reshape=input_data_as_numpy_array.reshape(+1,-1)
```

```python
# Prediction
prediction=model.predict(input_data_reshape)
print("Car price:",prediction)
```
